```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and it contains 'reservation_status_date' and 'is_canceled' columns.

# Extract month from the reservation date
df['month'] = df['reservation_status_date'].dt.month

# Create the count plot
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

# Adjust legend placement
ax1.legend(bbox_to_anchor=(1, 1))

# Set the plot title and labels
plt.title('Reservation Status Per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')

# Show the plot
plt.show()
```

```python
# Filter data for canceled bookings
cancelled_data = df[df['is_canceled'] == 1]

# Get the top 10 countries with the most cancellations
top_10_country = cancelled_data['country'].value_counts()[:10]

# Plot a pie chart
plt.figure(figsize=(6, 8))
plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)
plt.title("Top 10 Countries with Most Cancellations")
plt.show()
```

```python
df['month'] = pd.to_datetime(df['reservation_status_date']).dt.month
```

```python
month_mapping = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
                 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
df['month'] = df['month'].map(month_mapping)
```

```python
plt.figure(figsize=(15, 7))
plt.title('ADR per month', fontsize=30)

# Group data and prepare for plotting
adr_per_month = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()

# Plot
sns.barplot(x='month', y='adr', data=adr_per_month)
plt.xlabel('Month', fontsize=15)
plt.ylabel('ADR', fontsize=15)
plt.show()
```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date',inplace = True)

not_cancelled_data=df[df['is_canceled']== 0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize = (20,6))
plt.title('average daily rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr, label = 'not canceled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr, label = 'cancelled')
plt.legend()
```

```python
# Ensure reservation_status_date is in datetime format
cancelled_df_adr['reservation_status_date'] = pd.to_datetime(cancelled_df_adr['reservation_status_date'])
not_cancelled_df_adr['reservation_status_date'] = pd.to_datetime(not_cancelled_df_adr['reservation_status_date'])

# Plot ADR trends for canceled and non-canceled bookings
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate Over Time', fontsize=20)

plt.plot(
    not_cancelled_df_adr['reservation_status_date'], 
    not_cancelled_df_adr['adr'], 
    label='Not Canceled', 
    color='green'
)

plt.plot(
    cancelled_df_adr['reservation_status_date'], 
    cancelled_df_adr['adr'], 
    label='Canceled', 
    color='red'
)

# Add labels, legend, and grid
plt.xlabel('Reservation Status Date', fontsize=15)
plt.ylabel('Average Daily Rate (ADR)', fontsize=15)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
